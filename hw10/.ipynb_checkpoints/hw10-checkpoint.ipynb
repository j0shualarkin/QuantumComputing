```python
from qiskit import QuantumCircuit, ClassicalRegister, QuantumRegister, execute, Aer
from qiskit.visualization import plot_histogram
from math import pi
```

```python
backend = Aer.get_backend("qasm_simulator")
```

# Prepare the State $|\phi\rangle$

```python
phi_c = QuantumCircuit(2,2)
phi_c.h(0)
phi_c.cx(0,1)
phi_c.measure(range(2),range(2))
phi_c.draw('mpl')
```

# $|\phi\rangle = \frac{1}{\sqrt{2}}(|00\rangle + |11\rangle)$

```python
job_results = execute(phi_c,backend=backend,shots=1024).result()
counts = job_results.get_counts()
plot_histogram(counts)
```

# Experiment: $P_{\color{blue}\rightarrow\color{blue}\rightarrow}$

```python
xplus_xplus = QuantumCircuit(2,2)
xplus_xplus.h(0)
xplus_xplus.cx(0,1)
xplus_xplus.ry(0,0)
xplus_xplus.ry(0,1)
xplus_xplus.measure(range(2),range(2))
xplus_xplus.draw('mpl')
```

```python
job_results = execute(xplus_xplus,backend=backend,shots=1024).result()
counts = job_results.get_counts()
plot_histogram(counts)
```

# Mermin Circuit

```python
def mermin(theta_a,theta_b):
    c = QuantumCircuit(2,2)
    # prepare state 1/sqrt(2)(|00> + |11>)
    c.h(0)
    c.cx(0,1)
    # use negate on angles to reflect over x-axis
    # this allows us to measure in the standard basis
    c.ry(-theta_a,0)
    c.ry(-theta_b,1)
    c.measure(range(2),range(2))
    print(c.draw('mpl'))
    job_results = execute(c,backend=backend,shots=1024).result()
    counts = job_results.get_counts()
    return plot_histogram(counts)
```

```python
# The three angles we will use for our projectors
x_plus = 0
z_plus = 2*pi/3
y_plus = -z_plus
```

# Experiment $P_{\rightarrow\rightarrow}$  again

```python
mermin(0,0)
```

# Experiment $P_{\rightarrow\nwarrow}$

```python
mermin(x_plus,y_plus)
```

# Experiment $P_{\rightarrow\swarrow}$

```python
mermin(x_plus,z_plus)
```

# Experiment $P_{\nwarrow\nwarrow}$

```python
mermin(y_plus,y_plus)
```

# Experiment $P_{\nwarrow\swarrow}$

```python
mermin(y_plus,z_plus)
```

# Experiment $P_{\nwarrow\rightarrow}$

```python
mermin(y_plus,x_plus)
```

# Experiment $P_{\swarrow\swarrow}$ 

```python
mermin(z_plus,z_plus)
```

# Experiment $P_{\nwarrow\rightarrow}$

```python
mermin(z_plus,x_plus)
```

# Experiment $P_{\nwarrow\swarrow}$

```python
mermin(z_plus,y_plus)
```
